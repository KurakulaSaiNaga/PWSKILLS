### 1

`Matplotlib` is a popular Python library used for data visualization. It provides a wide range of tools for creating various 
types of plots and charts, making it a valuable tool for data scientists, analysts, and researchers. Matplotlib is highly 
customizable and can be used for creating publication-quality plots and figures. It is often used in conjunction with other 
Python libraries like NumPy and Pandas for data manipulation and analysis.

The `Pyplot` module of Matplotlib is a collection of functions that provide a simple and convenient way to create various types 
of plots and charts. 
Here are five common types of plots that can be created using Matplotlib's Pyplot module:
- Bar Plot
- Line Plot
- Histogram
- Pie Chart
- Scatter Plot

### 2

A scatter plot is a type of data visualization that is used to display individual data points as dots on a two-dimensional Cartesian coordinate system. Each dot on the scatter plot represents a single data point with two variables, typically referred to as the x-axis and y-axis variables. Scatter plots are used to visualize the distribution and relationships between these two variables and are particularly useful for identifying patterns, correlations, and outliers in data.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Random Numbers Plotting')
plt.scatter(x,y)
plt.show()
```

### 3

The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure. It allows you to arrange and display multiple plots in a grid or other customized layout. This can be particularly useful when you want to compare different datasets or visualize multiple aspects of your data side by side.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot 1')
plt.show()
```

### 4

A bar plot, also known as a bar chart or bar graph, is a type of data visualization that uses rectangular bars to represent and compare the values of different categories or groups. The length or height of each bar is proportional to the quantity or value it represents. Bar plots are commonly used to display and compare categorical data, making it easier to visualize and interpret the differences between categories or groups.

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
# Vertical
plt.bar(company,profit,color='blue')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Profits Of Companies')
plt.show()
# Horizontal
plt.barh(company,profit,color='green')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Profits Of Companies')
plt.show()
```

### 5

A box plot, also known as a whisker plot or box-and-whisker plot, is a data visualization technique that displays the distribution and summary statistics of a dataset. It is particularly useful for visualizing the spread and central tendency of numerical data, as well as identifying potential outliers.

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.random.normal(100, 10, 200)
plt.boxplot(a)
plt.show()
b=np.random.normal(90, 20, 200)
plt.boxplot(b)
plt.show()
```
