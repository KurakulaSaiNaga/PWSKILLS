In Bokeh, glyphs are visual markers or shapes that you can add to a plot to represent your data. Glyphs are used to display data points, lines, bars, and other geometric shapes in your plots. Bokeh provides a variety of glyph types that you can use to create different types of visualizations. Some common glyph types include circles, lines, squares, bars, and patches.

To add glyphs to a Bokeh plot, you typically use the glyph-specific functions provided by Bokeh's figure object (e.g., circle, line, square, etc.)

```python
from bokeh.plotting import figure, show, output_notebook

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Initialize a Bokeh figure
p = figure(title="Bokeh Glyphs Example", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add glyphs to the figure
p.circle(x, y, size=10, color="blue", alpha=0.5, legend_label="Circle Glyph")
p.line(x, y, line_width=2, color="green", legend_label="Line Glyph")
p.square(x, y, size=8, color="red", alpha=0.7, legend_label="Square Glyph")
p.vbar(x=x, top=y, width=0.3, color="purple", alpha=0.6, legend_label="Vertical Bar Glyph")

# Add a legend to the plot
p.legend.title = "Glyph Types"

# Specify the output mode (for Jupyter Notebook)
output_notebook()

# Display the plot
show(p)
```

A Bokeh server is a powerful feature of the Bokeh library that allows you to create interactive web applications with dynamic and real-time updates. With Bokeh server, you can build web applications that have interactive plots, dashboards, and widgets that respond to user input and update the visualizations in real time. This enables you to create data-driven and interactive web applications without the need for extensive web development expertise.

```python
from bokeh.plotting import curdoc, figure
from bokeh.models import ColumnDataSource
from bokeh.models.widgets import Slider

source = ColumnDataSource(data={'x': [1, 2, 3], 'y': [4, 5, 6]})
plot = figure()
plot.circle(x='x', y='y', source=source)

slider = Slider(start=1, end=10, value=1, step=1, title="Multiplier")
def update_data(attr, old, new):
    multiplier = slider.value
    new_data = {'x': [1, 2, 3], 'y': [4, 5, 6]}
    new_data['y'] = [y * multiplier for y in new_data['y']]
    source.data = new_data
slider.on_change('value', update_data)
curdoc().add_root(plot)
curdoc().add_root(slider)
#bokeh serve --show your_script.py
```
