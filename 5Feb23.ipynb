## 3

In Python, the __init__() function is a special method used in classes. It is called the constructor method and is automatically
executed when an object is created from a class. The primary purpose of the __init__() method is to initialize the attributes of
an object.

The __init__() method allows you to define and set the initial state of an object by accepting arguments and assigning them to 
the object's attributes. This method is useful when you want to ensure that certain attributes are initialized or set up before 
using the object.

Example-

class Person:
    
    def __init__(self, name, rollno):
        
        self.name = name
        
        self.rollno = rollno

    def display_info(self):
        
        print("Name:", self.name)
        
        print("RollNo:", self.rollno)
        
obj=Person('SaiNaga',42)

obj.display_info()

## 4

In object-oriented programming (OOP) in Python, the self parameter is used as a convention to refer to the instance of the class
within class methods. It acts as a reference to the instance itself and allows us to access its attributes and methods.

When defining methods within a class, including the __init__() method, the first parameter must be self. This parameter 
represents the instance of the class and is automatically passed when calling a method on an object.

The purpose of self is to differentiate between instance attributes (belonging to a specific object) and class attributes
(shared among all instances of the class). By using self, you can access and modify the attributes and methods of the current
instance.

## 5

Inheritance is a fundamental concept in object-oriented programming (OOP) that allows classes to inherit attributes and methods
from other classes. It promotes code reuse, modularity, and the hierarchical organization of classes. Inheritance enables the 
creation of new classes (derived or child classes) based on existing classes (base or parent classes), inheriting their 
properties while also adding new or modifying existing ones.

In Python, there are several types of inheritance:

#### Single inheritance:

In single inheritance, a derived class inherits from a single base class. It is the simplest form of inheritance.

class Animal:

    def eat(self):
    
        print("Animal is eating...")

class Dog(Animal):

    def bark(self):
    
        print("Dog is barking...")

dog = Dog()

dog.eat()  # Inherited from Animal

dog.bark()

#### Multiple inheritance: 

Multiple inheritance occurs when a derived class inherits from multiple base classes. It allows a class to inherit attributes
and methods from multiple sources.

class Mammal:

    def feed_milk(self):
    
        print("Mammal is feeding milk...")

class Reptile:

    def lay_eggs(self):
    
        print("Reptile is laying eggs...")

class Platypus(Mammal, Reptile):

    pass

platypus = Platypus()

platypus.feed_milk()  # Inherited from Mammal

platypus.lay_eggs()  # Inherited from Reptile

#### Multilevel inheritance:

Multilevel inheritance involves a chain of inheritance where a derived class becomes the base class for another derived class.

class Vehicle:

    def drive(self):
    
        print("Vehicle is being driven...")

class Car(Vehicle):

    def park(self):
    
        print("Car is parked...")

class SportsCar(Car):

    def race(self):
    
        print("Sports car is racing...")

sports_car = SportsCar()

sports_car.drive()  # Inherited from Vehicle

sports_car.park()  # Inherited from Car

sports_car.race()

#### Hierarchical inheritance:

Hierarchical inheritance occurs when multiple derived classes inherit from a single base class.

class Shape:

    def draw(self):
    
        print("Drawing shape...")

class Circle(Shape):

    def calculate_area(self):
    
        print("Calculating area of circle...")

class Rectangle(Shape):

    def calculate_area(self):
    
        print("Calculating area of rectangle...")

circle = Circle()

circle.draw()  # Inherited from Shape

circle.calculate_area()

rectangle = Rectangle()

rectangle.draw()  # Inherited from Shape

rectangle.calculate_area()
